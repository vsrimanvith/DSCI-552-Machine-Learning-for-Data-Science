<b>(f) Create a Python dictionary object whose keys are the headers of the dataframe
created in the read_csv() exercise and values are Python list objects that contain
data corresponding to the headers. (Here, use the iterrows() method to iterate
each row of the dataframe and copy it to a dictionary. However, there is an easier
way. Learn how the to_dict() method works by yourself later)</b>

```python
salaries_data_dict={}

for index, row in salaries_data.iterrows():
    for column_name, value in row.iteritems():
        if column_name in salaries_data_dict:
            salaries_data_dict[column_name].append(value)
        else:
            salaries_data_dict[column_name] = [value]
```

```python
salaries_data_dict = salaries_data.to_dict(orient='list')
```

<b>(b) Create multiple arrays and plot them with different styles, add legends, add text/mathematical equations on the plot.<br></b>

```python
x = np.linspace(0, 10, 100)
y1 = np.array([i for i in x])
y2 = np.sin(x)
y3 = np.cos(x)

plt.plot(x, y1, label='Linear: y = x', linestyle='--', color='green')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.text(2.5, 0.7, 'y = x', fontsize=12, color='green')
plt.grid()
plt.legend()
plt.show()


plt.plot(x, y2, label='sin(x)', linestyle='-', color='blue')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.text(2.5, 0.7, 'y = sin(x)', fontsize=12, color='blue')
plt.grid()
plt.legend()
plt.show()

plt.plot(x, y3, label='cos(x)', linestyle='--', color='red')
plt.title('Cosine Function')
plt.xlabel('x')
plt.ylabel('y')
plt.text(1.5, 0.2, 'y = cos(x)', fontsize=12, color='red')
plt.grid()
plt.legend()
plt.show()
```

<b>(c) Create multiple subplots, play around with the figure size, text font/size.<br></b>

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (14, 8))

axes[0, 0].plot(x, y1, label='sin(x)', linestyle='-', color='blue')
axes[0, 0].set_title('Sine Function', fontsize=16)
axes[0, 0].text(2.5, 0.8, 'y = sin(x)', fontsize=12, color='blue')
axes[0, 0].set_xlabel('x', fontsize=14)
axes[0, 0].set_ylabel('sin(x)', fontsize=14)

axes[0, 1].plot(x, y2, label='cos(x)', linestyle='--', color='red')
axes[0, 1].set_title('Cosine Function', fontsize=16)
axes[0, 1].text(1.5, 0.8, 'y = cos(x)', fontsize=12, color='red')
axes[0, 1].set_xlabel('x', fontsize=14)
axes[0, 1].set_ylabel('cos(x)', fontsize=14)

axes[1, 0].plot(x, y3, label='tan(x)', linestyle='-', color='green')
axes[1, 0].set_title('Tan Function', fontsize=16)
axes[1, 0].text(2.1, 1.5, 'y = tan(x)', fontsize=12, color='green')
axes[1, 0].set_xlabel('x', fontsize=14)
axes[1, 0].set_ylabel('tan(x)', fontsize=14)

axes[1, 1].plot(x, y4, label='exp(x)', linestyle='--', color='violet')
axes[1, 1].set_title('Exponential Function', fontsize=16)
axes[1, 1].text(1, 1, 'y = exp(x)', fontsize=12, color='violet')
axes[1, 1].set_xlabel('x', fontsize=14)
axes[1, 1].set_ylabel('exp(x)', fontsize=14)

axes[0, 0].grid(True)
axes[0, 1].grid(True)
axes[1, 0].grid(True)
axes[1, 1].grid(True)

plt.tight_layout()

plt.show()
```

<b>(d) Get familiar with get current axis (gca) handle to do the above tasks.<br></b>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

plt.sca(axes[0, 0])
ax1 = plt.gca()  
ax1.plot(x, y1, label='sin(x)', linestyle='-', color='blue')
ax1.set_title('Sine Function', fontsize=16, fontweight='bold')
ax1.text(2, 0.8, 'y = sin(x)', fontsize=12, color='blue')
ax1.grid(True)

plt.sca(axes[0, 1])
ax2 = plt.gca()
ax2.plot(x, y2, label='cos(x)', linestyle='--', color='red')
ax2.set_title('Cosine Function', fontsize=16, fontweight='bold')
ax2.text(2, 0.2, 'y = cos(x)', fontsize=12, color='red')
ax2.grid(True)

plt.sca(axes[1, 0])
ax3 = plt.gca()  
ax3.plot(x, y3, label='tan(x)', linestyle='-', color='brown')
ax3.set_title('Tan Function', fontsize=16, fontweight='bold')
ax3.text(2, 0.8, 'y = tan(x)', fontsize=12, color='brown')
ax3.grid(True)

plt.sca(axes[1, 1])
ax4 = plt.gca()  
ax4.plot(x, y4, label='exp(x)', linestyle='--', color='green')
ax4.set_title('Exponential Function', fontsize=16, fontweight='bold')
ax4.text(2, 0.5, 'y = exp(x)', fontsize=12, color='green')
ax4.grid(True)

for ax in axes.flatten():
    ax.set_xlabel('x', fontsize=14)
    ax.set_ylabel('y', fontsize=14)

plt.tight_layout()

plt.show()
```

<b>(e) Change the limits on x and y axes, use logarithmic axes to plot.</b>
