### (c) Convert the decision rules into a set of IF-THEN rules.

```python
text_representation = tree.export_text(DT_classifier, feature_names=col_names[0:6])
print(text_representation)
```

```python
def tree_to_code(tree, feature_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]
    feature_names = [f.replace(" ", "_")[:-5] for f in feature_names]
    print("def predict({}):".format(", ".join(feature_names)))

    def recurse(node, depth):
        indent = "    " * depth
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            print("{}if {} <= {}:".format(indent, name, np.round(threshold,2)))
            recurse(tree_.children_left[node], depth + 1)
            print("{}else:   if {} > {}".format(indent, name, np.round(threshold,2)))
            recurse(tree_.children_right[node], depth + 1)
        else:
            print("{}return {}".format(indent, tree_.value[node]))

    recurse(0, 1)
```

```python
def get_rules(tree, feature_names, class_names):
    tree_ = tree.tree_
    feature_name = [
        feature_names[i] if i != _tree.TREE_UNDEFINED else "undefined!"
        for i in tree_.feature
    ]

    paths = []
    path = []
    
    def recurse(node, path, paths):
        
        if tree_.feature[node] != _tree.TREE_UNDEFINED:
            name = feature_name[node]
            threshold = tree_.threshold[node]
            p1, p2 = list(path), list(path)
            p1 += [f"({name} <= {np.round(threshold, 3)})"]
            recurse(tree_.children_left[node], p1, paths)
            p2 += [f"({name} > {np.round(threshold, 3)})"]
            recurse(tree_.children_right[node], p2, paths)
        else:
            path += [(tree_.value[node], tree_.n_node_samples[node])]
            paths += [path]
            
    recurse(0, path, paths)

    # sort by samples count
    samples_count = [p[-1][1] for p in paths]
    ii = list(np.argsort(samples_count))
    paths = [paths[i] for i in reversed(ii)]
    
    rules = []
    for path in paths:
        rule = "if "
        
        for p in path[:-1]:
            if rule != "if ":
                rule += " and "
            rule += str(p)
        rule += " then "
        if class_names is None:
            rule += "response: "+str(np.round(path[-1][0][0][0],3))
        else:
            classes = path[-1][0][0]
            l = np.argmax(classes)
            rule += f"class: {class_names[l]} (proba: {np.round(100.0*classes[l]/np.sum(classes),2)}%)"
        rule += f" | based on {path[-1][1]:,} samples"
        rules += [rule]
        
    return rules
```

### (i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
def PCRCV(Xtrain,Ytrain,f, dim):
    k_f = KFold(f, shuffle=True, random_state=7)
    
    val_err = 0
    
    for train_idx, val_idx in k_f.split(Xtrain):
        # Train and Validation Split
        train_CV_X, val_CV_X = Xtrain.iloc[train_idx, :], Xtrain.iloc[val_idx, :]
        train_CV_Y, val_CV_Y = Ytrain.iloc[train_idx], Ytrain.iloc[val_idx]
        
        # Perform PCA
        pca = PCA(n_components = dim).fit(train_CV_X)
        pca_t = pca.transform(train_CV_X)
        pca_val = pca.transform(val_CV_X)
        
        # Perform Linear Regression
        model_reg = LinearRegression()
        model_reg.fit(pca_t, train_CV_Y)
        valid_pred = model_reg.predict(pca_val)
        valid_mse = mean_squared_error(val_CV_Y, valid_pred)
        
        # MSE Error for each iteration of fold
        val_err += valid_mse
        
    return val_err / f
```

```python
cv = pd.DataFrame(columns=['components', 'MSE'])

for i in tqdm(range(1, len(Xtrain.columns) + 1)):
    lst = []
    val_err = PCRCV(Xtrain,Ytrain, 5, i)
    lst.extend([i, val_err])
    lst_d = pd.Series(lst, index = ['components', 'MSE'])
    cv = cv.append(lst_d, ignore_index=True)

cv['components'] = cv['components'].astype(int)
    
cv
```

```python
# Find best possible component i.e. lowest MSE

best_comp = (cv[cv['MSE'] == cv['MSE'].min()]).values.tolist()[0]
best_comp

print('Minimum Validation MSE is {} When M = {}'.format(best_comp[1], int(best_comp[0])))
```

```python
# PCA model using best M
pca_bestM = PCA(n_components = int(best_comp[0])).fit(Xtrain)

# apply PCA to train and test set
train_pca_M = pca_bestM.transform(Xtrain)
test_pca_M = pca_bestM.transform(Xtest)

# linear regression with PCA results
model_reg_M = LinearRegression()
model_reg_M.fit(train_pca_M, Ytrain)
prediction_M = model_reg_M.predict(test_pca_M)
model_mse = mean_squared_error(Ytest, prediction_M)

# Storing Algorithm MSE
test_d = pd.Series(['PCA', model_mse], index = ['algorithm', 'MSE'])
algo_test_mse = algo_test_mse.append(test_d, ignore_index=True)

print('Test Error using PCA : ', model_mse)
```

### (j) In this section, we would like to fit a boosting tree to the data. As in classification trees, one can use any type of regression at each node to build a multivariate regression tree. Because the number of variables is large in this problem, one can use L1-penalized regression at each node. Such a tree is called L1 penalized gradient boosting tree. You can use XGBoost to fit the model tree. Determine α (the regularization term) using cross-validation.

```python
#pip install xgboost
```

```python
import xgboost as xgb
```

```python
# Calculate CV Score for selected alpha values
cv_res_xgb = pd.DataFrame(columns=['alpha', 'CV_Score'])

alphas = [10 ** i for i in range(-5, 5)]

# create k-fold cross validation splitter
k_f = KFold(5, shuffle=True, random_state=7)

for alpha in tqdm(alphas):
    lst = []
    x_model = xgb.XGBRegressor(objective="reg:squarederror", reg_alpha=alpha)
    cv_s = cross_val_score(x_model,Xtrain,Ytrain, cv=k_f, scoring="neg_mean_squared_error")
    lst.extend([alpha, np.abs(cv_s.mean())])
    lst_d = pd.Series(lst, index = ['alpha', 'CV_Score'])
    cv_res_xgb = cv_res_xgb.append(lst_d, ignore_index=True)
    
cv_res_xgb
```

```python
best_alpha_xg = (cv_res_xgb[cv_res_xgb['CV_Score'] 
                                == cv_res_xgb['CV_Score'].min()]).values.tolist()[0]

best_alpha_xg
print('Minimum Validation MSE is {} When Alpha = {}'.format(best_alpha_xg[1], float(best_alpha_xg[0])))
```

```python
# xgboost model with best alpha value
xg_model = xgb.XGBRegressor(objective='reg:squarederror', reg_alpha= float(best_alpha_xg[0]))
xg_model.fit(Xtrain,Ytrain)
pred = xg_model.predict(Xtest)
model_mse = mean_squared_error(Ytest, pred)

# Storing Algorithm MSE
test_d = pd.Series(['XGBoost', model_mse], index = ['algorithm', 'MSE'])
algo_test_mse = algo_test_mse.append(test_d, ignore_index=True)
 
print('Test Error using XGBoost : ', model_mse)
```
