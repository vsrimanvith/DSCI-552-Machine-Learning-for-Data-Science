### (b) Each instance has three labels: Families, Genus, and Species. Each of the labels has multiple classes. We wish to solve a multi-class and multi-label problem. One of the most important approaches to multi-label classification is to train a classifier for each label (binary relevance). We first try this approach:
#### i. Research exact match and hamming score/ loss methods for evaluating multi- label classification and use them in evaluating the classifiers in this problem

**Answer:** 

**Exact Match:** 
- Perfect match when the predicted labels exactly match the actual labels for a sample.
- Score of 1.0 indicates complete agreement.
- Score of 0.0 indicates no agreement between predicted and actual label sets.

**Hamming Loss:**
- Metric for multilabel classification evaluation.
- Measures average fraction of incorrectly predicted labels across all samples.
- Lower Hamming loss indicates better performance.
- Ideal score: 0.0, representing perfect predictions.

```python
def evaluate_multilabel_metrics(true_labels, predicted_labels):
    misclassifications = np.sum(true_labels.values != predicted_labels.values)
    hamming_loss = misclassifications / (true_labels.shape[0] * true_labels.shape[1])
    
    exact_match_ratio = np.sum(np.all(true_labels.values == predicted_labels.values, axis=1)) / true_labels.shape[0]
    
    hamming_loss = np.round(hamming_loss, 4)
    exact_match_ratio = np.round(exact_match_ratio, 4)
    
    print("Hamming Loss       :", hamming_loss)
    print("Exact Match Ratio  :", exact_match_ratio)
    return [hamming_loss, exact_match_ratio]
```

### ii. Train a SVM for each of the labels, using Gaussian kernels and one versus all classifiers. Determine the weight of the SVM penalty and the width of the Gaussian Kernel using 10 fold cross validation.1 You are welcome to try to solve the problem with both standardized 2 and raw attributes and report the results.

```python
def evaluate_multilabel(test_features, true_labels, classifiers):
    predicted_labels = pd.DataFrame(index=true_labels.index, columns=true_labels.columns)
    for label in true_labels.columns:
        model = classifiers[label]
        predicted_values = model.predict(test_features)
        predicted_labels.loc[:, label] = predicted_values
    return evaluate_multilabel_metrics(true_labels, predicted_labels)
```

```python
def find_best_model_and_evaluate(classifier, grid_search_params, train_features, train_labels, test_features, test_labels):
    model = GridSearchCV(estimator=classifier, **grid_search_params, n_jobs=-1)
    model.fit(train_features, train_labels)
    
    param_results = {key: [] for key in model.cv_results_['params'][0].keys()}
    for param in model.cv_results_['params']:
        for key, value in param.items():
            param_results[key].append(value)            
    means = np.round(np.array(model.cv_results_['mean_test_score']), 5)
    param_results['mean_test_score'] = means
    result_data = pd.DataFrame(param_results)
    display(result_data)

    print("\nThe best parameter setting is:")
    print(model.best_params_, "\n")
    print(classification_report(test_labels, model.predict(test_features)))
    
    return model
```

```python
svm_models = {}
evaluation_summary = {}
grid_search_params = {
    'scoring': 'f1_weighted',
    'param_grid': {
        'C': np.logspace(1, 4, 4), 
        'gamma': np.logspace(-3, 6, 10)
    }, 
    'cv': StratifiedKFold(10, random_state=5036, shuffle=True) 
}
method = "SVM with RBF Kernel without Standardization"
```

```python
labels = ['Family', 'Genus', 'Species']

for label in labels:
    print("Method:", method)
    print(f"Label: {label}")
    svm_models[label] = find_best_model_and_evaluate(SVC(kernel='rbf'), grid_search_params,
                                                     train_features, train_data[label], test_features, test_data[label])
```

```python
print("Evaluation Summary:", method)
evaluation_summary[method] = evaluate_multilabel(test_features, test_data.iloc[:, -4:-1], svm_models)
```

### iii. Repeat 1(b)ii with L1-penalized SVMs. Remember to standardize the attributes. Determine the weight of the SVM penalty using 10 fold cross validation.

```python
labels = ['Family', 'Genus', 'Species']

for label in labels:
    print("Method:", method_l1)
    print(f"Label: {label}")
    SVM_L1_classifiers[label] = find_best_model_and_evaluate(
        LinearSVC(penalty='l1', dual=False, max_iter=30000),
        grid_search_params_l1,
        train_features_std,
        train_data[label],
        test_features_std,
        test_data[label]
    )
```

```python
print("Evaluation Summary:", method_l1)
evaluation_summary[method_l1] = evaluate_multilabel(test_features_std, test_data.iloc[:, -4:-1], SVM_L1_classifiers)
```

### iv. Repeat 1(b)iii by using SMOTE or any other method you know to remedy class imbalance. Report your conclusions about the classifiers you trained.

```python
def smote_param_search(classifier, settings, train_x, train_y, test_x, test_y):
    smote_model = Pipeline([
        ('sampling', SMOTE()),
        ('classification', classifier)
    ])
    selected_model = find_best_model_and_evaluate(smote_model, settings, train_x, train_y, test_x, test_y)
    return selected_model
```

```python
SVC_smote_classifiers = {}
grid_search_params_smote = {
    'scoring': 'f1_weighted', 
    'param_grid': {'classification__C': np.logspace(1, 5, 10)}, 
    'cv': StratifiedKFold(10, random_state=5036, shuffle=True),
}
method_smote = "SVM with SMOTE"
```

```python
for label in labels:
    print("Method:", method_smote)
    print(f"Label: {label}")
    SVC_smote_classifiers[label] = smote_param_search(LinearSVC(penalty='l1', dual=False, max_iter=30000),
        grid_search_params_smote,
        train_features_std,
        train_data[label],
        test_features_std,
        test_data[label])
    
```

```python
print("Evaluation Summary:", method_smote)
evaluation_summary[method_smote] = evaluate_multilabel(test_features_std, test_data.iloc[:, -4:-1], SVC_smote_classifiers)
```

# 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set

## Monte-Carlo Simulation: Perform the following procedures 50 times, and report the average and standard deviation of the 50 Hamming Distances that you calculate.

### (a) Use k-means clustering on the whole Anuran Calls (MFCCs) Data Set (do not split the data into train and test, as we are not performing supervised learning in this exercise). Choose k ∈ {1, 2, . . . , 50} automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any other method you know.
### (b) In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species.
### (c) Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance, Hamming score, and Hamming loss between the true labels and the labels assigned by clusters.

```python
datasetX = data
datasetY = datasetX[['Family', 'Genus', 'Species']]
datasetX = datasetX.drop(['Family', 'Genus', 'Species'], axis=1)
hamming_loss_list = []
for experiment_no in range(1,51):
    print('Experiment ', experiment_no)
    experimentSilouhetteScoreDict = {}
    for num_clusters in range(2,51):
        kmc = KMeans(n_clusters = num_clusters, random_state=2022)
        kmc.fit(datasetX)
        experimentSilouhetteScoreDict[num_clusters] = silhouette_score(datasetX, kmc.fit_predict(datasetX))
    for key,value in experimentSilouhetteScoreDict.items():
        print('k : ', key, 'Silouhette Scores : ', value)
    bestKValue = max(experimentSilouhetteScoreDict, key=experimentSilouhetteScoreDict.get)
    finalKmm = KMeans(n_clusters = bestKValue, random_state=2022)
    cluster = pd.DataFrame(finalKmm.fit_predict(datasetX), columns=['label'])
    fullDataConcatenated = pd.concat([data,cluster], axis=1)
    mostPopulousClassDict = {}
    for i in range(0, bestKValue):
        currdf = fullDataConcatenated[fullDataConcatenated.label == i]
        mostPopulousClassDict[i] = [currdf['Family'].value_counts().idxmax(),currdf['Genus'].value_counts().idxmax(),currdf['Species'].value_counts().idxmax()]
    print('Most populous classes per cluster', mostPopulousClassDict)
    datasetPredicted = fullDataConcatenated
    datasetPredicted['kmm-family'] = None
    datasetPredicted['kmm-genus'] = None
    datasetPredicted['kmm-species'] = None
    for i in range(0, bestKValue):
        datasetPredicted['kmm-family'] = np.where(i==datasetPredicted['label'], mostPopulousClassDict[i][0], datasetPredicted['kmm-family'])
        datasetPredicted['kmm-genus'] = np.where(i==datasetPredicted['label'], mostPopulousClassDict[i][1], datasetPredicted['kmm-genus'])
        datasetPredicted['kmm-species'] = np.where(i==datasetPredicted['label'], mostPopulousClassDict[i][2], datasetPredicted['kmm-species'])
    hamming_for_current_experiment = []
    hamming_for_current_experiment.append(hamming_loss(datasetPredicted['kmm-family'],datasetPredicted['Family']))
    hamming_for_current_experiment.append(hamming_loss(datasetPredicted['kmm-genus'],datasetPredicted['Genus']))
    hamming_for_current_experiment.append(hamming_loss(datasetPredicted['kmm-species'],datasetPredicted['Species']))
    fpredicted = datasetPredicted['kmm-family'].values.tolist()
    factual = datasetPredicted['Family'].values.tolist()
    gpredicted = datasetPredicted['kmm-genus'].values.tolist()
    sactual = datasetPredicted['Species'].values.tolist()
    spredicted = datasetPredicted['kmm-species'].values.tolist()
    gactual = datasetPredicted['Genus'].values.tolist()
    family_ham_dist = 0
    genus_ham_dist = 0
    species_ham_dist = 0
    for i in range(len(fpredicted)):
        if fpredicted[i]==factual[i]:
            family_ham_dist+=1
        if gpredicted[i]==gactual[i]:
            genus_ham_dist+=1
        if spredicted[i]==sactual[i]:
            species_ham_dist+=1
    print('Hamming distance for Family', family_ham_dist)
    print('Hamming distance for Genus', genus_ham_dist)
    print('Hamming distance for Species', species_ham_dist)
    print('Average Hamming Loss for Family', hamming_for_current_experiment[0])
    print('Average Hamming Loss for Genus', hamming_for_current_experiment[1])
    print('Average Hamming Loss for Species', hamming_for_current_experiment[2])
    print('Average Hamming Score for Family', 1-hamming_for_current_experiment[0])
    print('Average Hamming Score for Genus', 1-hamming_for_current_experiment[1])
    print('Average Hamming Score for Species', 1-hamming_for_current_experiment[2])
    hamming_loss_list.append((hamming_for_current_experiment[0]+hamming_for_current_experiment[1]+hamming_for_current_experiment[2])/3)
    print('--------------------------------------------------------')
    
print('Mean of all 50 experiments', np.mean(hamming_loss_list))
print('Standard Deviation of all 50 experiments', np.std(hamming_loss_list))
```
