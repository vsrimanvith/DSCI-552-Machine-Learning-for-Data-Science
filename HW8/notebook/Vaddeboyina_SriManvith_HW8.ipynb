```python
def semi_supervised(trainX_norm, trainY, kwargs): 
    labeled_idx = []
    unlabeled_idx = []
    
    for c in [0, 1]:
        c_label_idx, = np.where(trainY == c)
        labeled_idx.extend(np.random.choice(c_label_idx, round(len(c_label_idx) * 0.5), replace = False))
    
    for i in range(len(trainX_norm)):
        if i not in labeled_idx:
            unlabeled_idx.append(i)
    
    labeled_x = trainX_norm.iloc[labeled_idx].to_numpy().tolist()
    labeled_y = trainY.iloc[labeled_idx].to_numpy().tolist()
    unlabeled_x = trainX_norm.iloc[unlabeled_idx].to_numpy().tolist()
    unlabeled_y = trainY.iloc[unlabeled_idx].to_numpy().tolist()
    
    model = GridSearchCV(estimator=LinearSVC(penalty='l1', dual=False), **kwargs)
    model.fit(labeled_x, labeled_y)
    
    while len(unlabeled_x) > 0:
        scores = model.decision_function(unlabeled_x)
        sorted_idx = sorted([i for i in range(len(unlabeled_x))], key = lambda x : abs(scores[x]))
        unlabeled_x = [unlabeled_x[i] for i in sorted_idx]
        unlabeled_y = [unlabeled_y[i] for i in sorted_idx]
        labeled_x.append(unlabeled_x.pop())
        labeled_y.append(unlabeled_y.pop())
        model.best_estimator_.fit(labeled_x, labeled_y) 
    return model


kwargs = {
        'param_grid': {'C': np.logspace(-2, 7, 10)},
        'scoring': 'f1_weighted',
        'n_jobs': -1
    }
sum_train_accuracy, sum_train_precision, sum_train_recall, sum_train_f1score, sum_train_auc = 0,0,0,0,0
sum_test_accuracy, sum_test_precision, sum_test_recall, sum_test_f1score, sum_test_auc = 0,0,0,0,0
    
for k in range(30):
    np.random.seed(k)
    
    B_idx = np.where(df_wdbc['Diagnosis'] == 0)[0]
    M_idx = np.where(df_wdbc['Diagnosis'] == 1)[0]

    B_test_idx = np.random.choice(B_idx, round(len(B_idx) * 0.2), replace=False)
    M_test_idx = np.random.choice(M_idx, round(len(M_idx) * 0.2), replace=False)

    test_idx = np.concatenate([B_test_idx, M_test_idx])
    train_idx = np.setdiff1d(np.arange(len(df_wdbc)), test_idx)

    trainX, trainY = df_wdbc.iloc[train_idx, 1:].copy(), df_wdbc.iloc[train_idx, 0].copy()
    testX, testY = df_wdbc.iloc[test_idx, 1:].copy(), df_wdbc.iloc[test_idx, 0].copy()

    trainX_norm, testX_norm = Normalizer().fit_transform(trainX), Normalizer().fit_transform(testX)
    
    trainX_norm, testX_norm = pd.DataFrame(trainX_norm), pd.DataFrame(testX_norm)

    kwargs['cv'] = StratifiedKFold(5, random_state = i, shuffle = True)
    
    model = semi_supervised(trainX_norm, trainY, kwargs)
    
    y_pred_train_prob = model.decision_function(trainX_norm)
    y_pred_train = model.predict(trainX_norm)
    train_model_results = classification_report(trainY, y_pred_train, output_dict=True)
    train_accuracy = train_model_results['accuracy']
    train_precision = train_model_results['weighted avg']['precision']
    train_recall = train_model_results['weighted avg']['recall']
    train_f1score = train_model_results['weighted avg']['f1-score']
    fpr_train, tpr_train, _ = roc_curve(trainY, y_pred_train_prob, pos_label = 1)
    train_auc = auc(fpr_train, tpr_train)

    y_pred_test_prob = model.decision_function(testX_norm)
    y_pred_test = model.predict(testX_norm)
    test_model_results = classification_report(testY, y_pred_test,output_dict=True)
    test_accuracy = test_model_results['accuracy']
    test_precision = test_model_results['weighted avg']['precision']
    test_recall = test_model_results['weighted avg']['recall']
    test_f1score = test_model_results['weighted avg']['f1-score']
    fpr_test, tpr_test, _ = roc_curve(testY, y_pred_test_prob, pos_label = 1)
    test_auc = auc(fpr_test, tpr_test)

    print('Iteration:',k+1)
    print("----------------")
    print("Train Accuracy:",train_accuracy)
    print("Test Accuracy:",test_accuracy)
    print("Best Params:",model.best_params_)
    print()
    print()
    
    sum_train_accuracy += train_accuracy
    sum_train_precision += train_precision
    sum_train_recall += train_recall
    sum_train_f1score += train_f1score
    sum_train_auc += train_auc
    
    sum_test_accuracy += test_accuracy
    sum_test_precision += test_precision
    sum_test_recall += test_recall
    sum_test_f1score += test_f1score
    sum_test_auc += test_auc


print("\033[1mResults:")
print("-----------")
print()
print("\033[1mTrain Data:")
print("-------------")
print("Average Accuracy:", sum_train_accuracy/30)
print("Average Precision:", sum_train_precision/30)
print("Average Recall:", sum_train_recall/30)
print("Average F1-score:", sum_train_f1score/30)
print("Average AUC score:", sum_train_auc/30)
print("Average Confusion Matrix:")
print(confusion_matrix(trainY, y_pred_train))
print()

fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(fpr_train, tpr_train, label='ROC curve (Area = %0.4f)' % train_auc)
axes.plot([0, 1], [0, 1], "--")
axes.set_xlabel("False Positive Rate (FPR)")
axes.set_ylabel("True Positive Rate (TPR)")
axes.set_title("ROC curve of train set")
axes.legend()
plt.show()


print("\033[1mTest Data:")
print("-------------")
print("Average Accuracy:", sum_test_accuracy/30)
print("Average Precision:", sum_test_precision/30)
print("Average Recall:", sum_test_recall/30)
print("Average F1-score:", sum_test_f1score/30)
print("Average AUC score:", sum_test_auc/30)
print("Confusion Matrix:")
print(confusion_matrix(testY, y_pred_test))
print()

fig, axes = plt.subplots(figsize=(8, 6))
axes.plot(fpr_test, tpr_test, label='ROC curve (Area = %0.4f)' % test_auc)
axes.plot([0, 1], [0, 1], "--")
axes.set_xlabel("False Positive Rate (FPR)")
axes.set_ylabel("True Positive Rate (TPR)")
axes.set_title("ROC curve of test set")
axes.legend()
plt.show()  
print()
```

```python
def get_distance(centers, X, cluster_label):
    dist = np.ndarray((X.shape[0],))
    proba = np.ndarray((X.shape[0], 2))
    for i, x in enumerate(X):
        center = centers[cluster_label[i]]
        dist[i] = np.linalg.norm(x - center)
        proba[i, 0] = np.linalg.norm(x - centers[0])
        proba[i, 1] = np.linalg.norm(x - centers[1])
    
    proba = softmax(proba, axis=1).astype(np.float64)
    return dist, proba

def get_majority(cluster_label, dist, Y):
    label2cluster = np.ndarray((2,), dtype='int')
    major_label = cluster_label.copy()
    for cluster in [0, 1]:
        sample_idx, = np.where(cluster_label == cluster)
        top30 = sorted(sample_idx.tolist(), key=lambda i : dist[i])[:30]
        majority = Y.iloc[top30].value_counts().index[0]
        major_label[sample_idx] = majority
        label2cluster[majority] = cluster
    return label2cluster, major_label
```

```python
kwargs = {
    'param_grid' : {'C' : np.logspace(-2, 7, 20)}, 
    'scoring' : 'f1_weighted',
    'n_jobs' : -1
}

active_learning_curve = []   
for i in range(50):
    test_idx = np.random.choice(train_data.index, size=472, replace=False)
    train_idx = np.setdiff1d(train_data.index, test_idx)

    train = train_data.loc[train_idx]
    test = train_data.loc[test_idx]
    trainX, trainY = train.iloc[:, :-1], train.iloc[:, -1]
    testX, testY = test.iloc[:, :-1], test.iloc[:, -1]

    trainX_norm, testX_norm = Normalizer().fit_transform(trainX), Normalizer().fit_transform(testX)
    trainX_norm, testX_norm = pd.DataFrame(trainX_norm), pd.DataFrame(testX_norm)
    trainX,testX,trainY,testY = trainX_norm,testX_norm,trainY,testY
    kwargs['cv'] = StratifiedKFold(5, random_state = i, shuffle = True)
    
    all_idx = np.arange(trainX.shape[0])
    new_idx, accuracy = [], []
    np.random.shuffle(all_idx)
    pool, all_idx = all_idx[:10], all_idx[10:]
    new_idx.extend(pool)
    
    while len(new_idx) <= 900:
        new_train_x, new_train_y = trainX.iloc[new_idx], trainY.iloc[new_idx]
        model = GridSearchCV(estimator = LinearSVC(penalty = 'l1', dual = False), **kwargs)
        model.fit(new_train_x, new_train_y)
        predictions = model.predict(testX)
        result = classification_report(testY, predictions, output_dict = True)
        accuracy.append(result['accuracy'])
        
        if len(all_idx) == 0:
            break
        
        dist = model.decision_function(trainX.iloc[all_idx])
        sortedarg = np.arange(len(all_idx))
        sorted(sortedarg, key=lambda i : abs(dist[i]))
        all_idx = all_idx[sortedarg]
        pool, all_idx = all_idx[:10], all_idx[10:]
        new_idx.extend(pool)
    
    active_learning_curve.append(accuracy)
    print('Iteration:', i+1)
    print("----------------")
    print("Train Accuracy:",accuracy[0])
    print("Test Accuracy:",accuracy[-1])
    print()
    print()
active_learning_curve = np.mean(np.array(active_learning_curve), axis = 0)
```
