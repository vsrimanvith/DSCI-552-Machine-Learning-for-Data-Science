```python
X_train = column_2c_train_set.iloc[:, :-1]
y_train = column_2c_train_set.iloc[:, -1]

X_test = column_2c_test_set.iloc[:, :-1]
y_test = column_2c_test_set.iloc[:, -1]
```

### (c) Classification

i. Write code for k-nearest neighbors with Euclidean metric (or use a software
package).

#### i. Euclidean Metric

```python
def KNN_Algorithm(X_train, y_train, X_test, y_test, k, k_dict, metric, metric_params, weights,p):
    KNN_model = KNeighborsClassifier(n_neighbors = k, metric = metric, metric_params = metric_params,weights = weights,p = p)
    KNN_model.fit(X_train, y_train)

    k_error_dict = {}
    k_error_dict['train'] = 1 - KNN_model.score(X_train, y_train)
    k_error_dict['test'] = 1 - KNN_model.score(X_test, y_test)
    k_dict[k] = k_error_dict
    
def KNN_Algorithm_(X_train, y_train, X_test, y_test, k, metric, metric_params):
    KNN_model = KNeighborsClassifier(n_neighbors = k, metric = 'euclidean')
    KNN_model.fit(X_train, y_train)
    y_pred = KNN_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred)

    return test_error
```

iii. Since the computation time depends on the size of the training set, one may only use a subset of the training set. Plot the best test error rate , 3 which is obtained by some value of k , against the size of training set, when the size of training set is N  { 10 , 20 , 30 , . . . , 210 } . 4 Note: for each N , select your training set by choosing the first b N/ 3 c rows of Class 0 and the first N  b N/ 3 c rows of Class 1 in the training set you created in 1(b)iii. Also, for each N , select the optimal k from a set starting from k = 1, increasing by 5. For example, if N = 200, the optimal k is selected from { 1 , 6 , 11 , . . . , 196 } . This plot is called a Learning Curve

#### iii. Learning Curve

```python
optimal_k = []
optimal_errors = []

N_values = list(range(10, 211, 10))

for N_value in N_values:
    n_class_0 = N_value // 3
    n_class_1 = N_value - n_class_0

    column_2c_train_set_subset = pd.concat([
        column_2c_train_set[column_2c_train_set["label"] == 0].iloc[:n_class_0],
        column_2c_train_set[column_2c_train_set["label"] == 1].iloc[:n_class_1]])
    
    X_train_ = column_2c_train_set_subset.iloc[:, :-1]
    y_train_ = column_2c_train_set_subset.iloc[:, -1]
    
    best_test_error_rate = float("inf")
    best_k_value = 0
    
    for k_value in range(1, N_value, 5):
        test_error = KNN_Algorithm_(X_train_, y_train_, X_test, y_test,k_value,'euclidean',{})

        if test_error < best_test_error_rate:                
            best_k_value = k_value
            best_test_error_rate = test_error

    optimal_k.append(best_k_value)
    optimal_errors.append(best_test_error_rate)

figure(figsize=(10, 6))

plt.plot(N_values, optimal_errors, label="Best test error rate")
plt.title("Learning Curve")
plt.xlabel("Size of training set - N")
plt.ylabel("Best test error rate")
plt.legend()
plt.show()
```

### (d) Other Metrics

Replace the Euclidean metric with the following metrics and test them. Summarize the test errors (i.e., when k = k) in a table. Use all of your training data and select the best k when { 1 , 6 , 11 , . . . , 196 }

```python
metrics_dataframe = pd.DataFrame(columns=["metric", "best k value", "test error"])
```

#### i. Minkowski Distance.

##### A. Manhattan Distance with p = 1.

```python
k_dict = {}

k_values = list(range(1, 197, 5))

for k_value in k_values:
    KNN_Algorithm(X_train, y_train, X_test, y_test, k_value, k_dict,'minkowski',{},'uniform',1)

manhattan_min_test_error_k_value = min(k_dict, key = lambda k: k_dict[k]['test'])
best_k_values_minkowski = [k for k, v in k_dict.items() if v['test'] == k_dict[manhattan_min_test_error_k_value]['test']]


train_errors = [values['train'] for values in k_dict.values()]
test_errors = [values['test'] for values in k_dict.values()]
min_test_error_minkowski = min(test_errors)

figure(figsize=(12, 5))
plt.plot(k_values, test_errors, label = "testing error")
plt.title("k value vs error rate")
plt.xlabel("k value")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
print("Most suitable k* value: "+ str(best_k_values_minkowski)+",", "Test error: " + str(min_test_error_minkowski))
minkowski_data = {"metric": 'manhattan (minkowski, p = 1)', "best k value": best_k_values_minkowski, "test error": min_test_error_minkowski}
metrics_dataframe = metrics_dataframe.append(minkowski_data, ignore_index=True)
```

##### B. With log10(p) in {0.1, 0.2, 0.3, ... ,1}.

```python
test_errors = []

p_values = list(np.logspace(0.1, 1, 10))

best_test_error_rate = float("inf")

for p_value in p_values:
    KNN_model = KNeighborsClassifier(n_neighbors = best_k_values_minkowski[0], metric = 'minkowski', p = p_value)
    KNN_model.fit(X_train, y_train)
    y_pred_ = KNN_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred_)
    test_errors.append(test_error)

    if test_error < best_test_error_rate:                
        best_p_value = p_value
        best_test_error_rate = test_error

figure(figsize=(12, 5))
plt.plot(np.linspace(0.1, 1, 10), test_errors, label = "testing error")

plt.title("p value vs test error")
plt.xlabel("p value")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
print("Most suitable log10(p) value: "+ str(np.log10(best_p_value))+",", "Test error: " + str(best_test_error_rate))
minkowski_logp_data = {"metric": 'minkowski (log10(p) = 0.6)', "best k value": best_k_values_minkowski[0], "test error": best_test_error_rate}
metrics_dataframe = metrics_dataframe.append(minkowski_logp_data, ignore_index=True)
```

##### C. Chebyshev Distance With p -> infinity.

```python
k_dict = {}

k_values = list(range(1, 197, 5))

for k_value in k_values:
    KNN_Algorithm(X_train, y_train, X_test, y_test, k_value, k_dict,'chebyshev',{},'uniform',1)

min_test_error_k_value_chebyshev = min(k_dict, key = lambda k: k_dict[k]['test'])
min_test_error_value_chebyshev = k_dict[min_test_error_k_value_chebyshev]['test']


train_errors = [values['train'] for values in k_dict.values()]
test_errors = [values['test'] for values in k_dict.values()]

figure(figsize=(12, 5))
plt.plot(k_values, test_errors, label = "testing error")
plt.title("k value vs error rate")
plt.xlabel("k value")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
print("Most suitable k* value: "+ str(min_test_error_k_value_chebyshev)+",", "Test error: " + str(min_test_error_value_chebyshev))
minkowski_chebyshev_data = {"metric": 'chebyshev (minkowski, p = inf)', "best k value": min_test_error_k_value_chebyshev, "test error": min_test_error_value_chebyshev}
metrics_dataframe = metrics_dataframe.append(minkowski_chebyshev_data, ignore_index=True)
```

#### ii. Mahalanobis Distance.

```python
test_errors = []

k_dict = {}
best_test_error_rate = float("inf")

k_values = list(range(1, 200, 5))

for k_value in k_values:
    KNN_model = KNeighborsClassifier(n_neighbors = k_value, metric = 'mahalanobis', weights='uniform',metric_params={"VI": np.linalg.pinv(np.cov(X_train, rowvar=False))})
    KNN_model.fit(X_train, y_train)
    y_pred_ = KNN_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred_)
    test_errors.append(test_error)
    
    if test_error < best_test_error_rate:                
        best_k_value = k_value
        best_test_error_rate = test_error
        
matching_k_values_mahalanobis = [k for k, error in zip(k_values, test_errors) if error == best_test_error_rate]

figure(figsize=(12, 5))
plt.plot(k_values, test_errors, label = "testing error")
plt.title("k value vs error rate")
plt.xlabel("k value")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
print("Most suitable k* value: "+ str(matching_k_values_mahalanobis)+",", "Test error: " + str(best_test_error_rate))
mahalanobis_data = {"metric": 'mahalanobis', "best k value": matching_k_values_mahalanobis, "test error": best_test_error_rate}
metrics_dataframe = metrics_dataframe.append(mahalanobis_data, ignore_index=True)
```
