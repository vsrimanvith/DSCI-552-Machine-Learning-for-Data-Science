In the univariate realm, missing values are replaced with fixed values, such as the mean, median, or the most frequently occurring value within the same feature.
```python
column_stats = {}
for column_name in imputed_X_train.columns:
    column_mean = imputed_X_train[column_name].mean()
    column_std = imputed_X_train[column_name].std()
    column_cv = column_std / column_mean
    column_stats[column_name] = [column_mean, column_std, column_cv]

column_stats_df = pd.DataFrame(column_stats)

column_stats_df.index = ['Mean', 'Standard Deviation', 'Coefficient of Variation']

column_stats_df = column_stats_df.T
column_stats_df
```

### (c) Train a random forest
Train a random forest to classify the data set. Do NOT compensate for class
imbalance in the data set. Calculate the confusion matrix, ROC, AUC, and
misclassification for training and test sets and report them (You may use pROC
package). Calculate Out of Bag error estimate for your random forset and compare

```python
model_rf = RandomForestClassifier(max_depth=5, random_state = 42, oob_score=True)
model_rf.fit(imputed_X_train, y_train)

y_pred_train = model_rf.predict(imputed_X_train)
y_pred_train_prob = model_rf.predict_proba(imputed_X_train)
fpr, tpr, thresholds = roc_curve(y_train, y_pred_train_prob[:, 1], pos_label='pos')
roc_auc = auc(fpr, tpr)

print("\033[1mConfusion Matrix:", confusion_matrix(y_train, y_pred_train),sep="\n")
print()
print("\033[1mTrain accuracy:", accuracy_score(y_train, y_pred_train))
print("\033[1mTrain error:", 1-accuracy_score(y_train, y_pred_train))

fig, axes = plt.subplots(figsize=(10, 6))
axes.plot(fpr, tpr, label='ROC curve (Area = %0.4f)' % roc_auc)
axes.plot([0, 1], [0, 1], "--")
axes.set_xlabel("False Positive Rate (FPR)")
axes.set_ylabel("True Positive Rate (TPR)")
axes.set_title("ROC curves of train set")
axes.legend()
plt.show()
```

```python
y_pred_test = model_rf.predict(imputed_X_test)
y_pred_test_prob = model_rf.predict_proba(imputed_X_test)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_test_prob[:, 1], pos_label='pos')
roc_auc = auc(fpr, tpr)

print("\033[1mConfusion Matrix:", confusion_matrix(y_test, y_pred_test),sep="\n")
print()
print("\033[1mTest accuracy:", accuracy_score(y_test, y_pred_test))
print("\033[1mTest error:", 1-accuracy_score(y_test, y_pred_test))

fig, axes = plt.subplots(figsize=(10, 6))
axes.plot(fpr, tpr, label='ROC curve (Area = %0.4f)' % roc_auc)
axes.plot([0, 1], [0, 1], "--")
axes.set_xlabel("False Positive Rate (FPR)")
axes.set_ylabel("True Positive Rate (TPR)")
axes.set_title("ROC curves of test set")
axes.legend()
plt.show()
```

```python
oob_train_err = 1 - model_rf.oob_score_
print("\033[1mTrain Out of Bag Error : ", oob_train_err)
print("\033[1mTest error             : ", 1 - accuracy_score(y_test, y_pred_test))
```

## Discussion
The close proximity between the OOB error estimate and the test error rate implies that the random forest model is effectively maintaining its accuracy on previously unseen data. This observation underscores the model's robustness and suggests it is not prone to overfitting.

### (e) XGBoost and Model Trees

In the case of a univariate tree, only one input dimension is used at a tree split.
In a multivariate tree, or model tree, at a decision node all input dimensions can be used and thus it is more general. In univariate classification trees, majority polling is used at each node to determine the split of that node as the decision rule. In model trees, a (linear) model that relies on all of the variables is used to determine the split of that node (i.e. instead of using $X_j > s$ as the decision rule, one has $\sum_{j}\beta_{j}X_j > s$. as the decision rule). Alternatively, in a regression tree, instead of using average in the region associated with each node, a linear regression model is used to determine the value associated with that node. One of the methods that can be used at each node is Logistic Regression. Because the number of variables is large in this problem, one can use $L1$ penalized logistic regression at each node. ## 4. ISLR 8.4.5

<b>Suppose we produce ten bootstrapped samples from a data set
containing red and green classes. We then apply a classifcation tree
to each bootstrapped sample and, for a specifc value of X, produce
10 estimates of P(Class is Red|X):<br><br></b>
<b><center>0.1, 0.15, 0.2, 0.2, 0.55, 0.6, 0.6, 0.65, 0.7, and 0.75.</center><br>
There are two common ways to combine these results together into a
single class prediction. One is the majority vote approach discussed in
this chapter. The second approach is to classify based on the average
probability. In this example, what is the fnal classifcation under each
of these two approaches?</b>

<b>Answer:</b> In the context of classification based on probability thresholds, two common techniques are used:

1. Majority Polling: This approach involves classifying data points by looking at their associated probabilities. When the probability of an event is greater than or equal to 0.5, it is classified as true; otherwise, it's classified as false. For example, out of 10 samples, if 4 have probabilities below 0.5 and 6 have probabilities greater than or equal to 0.5, the majority class determines the classification. In this case, the majority class is the positive or true class, resulting in a classification of "red."

2. Average Probability: In this method, the classification is determined by calculating the mean of the given probabilities. If the average probability is greater than or equal to 0.5, the classification will be true; otherwise, it will be false. For instance, if the mean probability is 0.45, the final prediction is "green."

These techniques offer different ways to classify data based on probabilities, with Majority Polling focusing on the majority class and Average Probability using the mean probability to make predictions. The choice between these methods depends on the specific problem and the desired classification outcome.

```python
data = pd.DataFrame({'X1': [3, 2, 4, 1, 2, 4, 4],
                     'X2': [4, 2, 4, 4, 1, 3, 1],
                     'Y': ['red', 'red', 'red', 'red', 'blue', 'blue', 'blue']})

X, y = data[['X1', 'X2']], data['Y']

model = svm.SVC(kernel = 'linear', C = 15000)
model.fit(X, y)

fig, ax = plt.subplots(1, 2, figsize = (18, 6))
sns.scatterplot(data = data, x = 'X1', y = 'X2', hue = 'Y', ax = ax[0])
ax[1].scatter(data['X1'], data['X2'], c = data['Y'], s = 30, cmap = plt.cm.Paired)

xlim, ylim = ax[1].get_xlim(), ax[1].get_ylim()

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 30), np.linspace(ylim[0], ylim[1], 30))
zz = model.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

ax[1].contour(xx, yy, zz, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
ax[1].scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1], s=100, linewidth=1, facecolors='none', edgecolors='k',label = "Support Vectors")
ax[1].contour(xx, yy, zz - 0.2, colors='c', levels=[0])
ax[1].scatter([4.0], [2.0], color='r', marker='^', s=50, label = "Additional point")

print("\033[1mIntercept    :", model.intercept_)
print("\033[1mCoefficients :", model.coef_)

plt.legend()
plt.show()
```
